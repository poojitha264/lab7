```python
data=pd.read_csv('/content/Japan_life_expectancy.csv')
```

```python
data1=data.drop('Prefecture',axis=1)
```

```python
data1
```

```python
y=data1['Life_expectancy']
y
```

```python
x=data1.drop('Life_expectancy',axis=1)
x
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
k_values = list(range(1, 21))
mse_values = []
```

```python
for k in k_values:
      knn = KNeighborsRegressor(n_neighbors=k)
      knn.fit(x_train, y_train)
      y_pred = knn.predict(x_test)
      mse = mean_squared_error(y_test,y_pred)
      mse_values.append(mse)
print(mse_values)
```

```python
plt.plot(k_values, mse_values, marker='o', linestyle='-')
plt.title('Mean Squared Error vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Squared Error')
plt.show()
#for k=1 the mean squared error is maximum &k=14 the mean squared error is minimum
```
